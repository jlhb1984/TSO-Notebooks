```python
import pandas as pd
import numpy as np
```

```python
ubd_nf=pd.read_csv('/home/jlhb1984/py-projects/data/cases/crm_ubd10122025.csv',sep=';',encoding='latin1')
ubd_nf['Last Event Date'].fillna('Empty',inplace=True)
ubd=ubd_nf[ubd_nf['Last Event Date'].str.contains('2025')]
row_number=ubd.shape[0]
ubd.info()
#ubd.info()
```

```python
ubd['Dealer'].value_counts()
```

```python
vec_aux=[]
for i in range(0,row_number):
    aux=ubd.iloc[i,14].replace(',','.')
    vec_aux.append(float(aux))
ubd['Dec_Monthly_fee']=vec_aux

ubd.info()
#ubd.to_csv('/home/jlhb1984/py-projects/data/cases/mrr.csv')
```

```python
tso_mobile_colombia=ubd[ubd['Dealer'].str.contains('TSO Mobile - Colombia')]
tso_mobile_peru=ubd[ubd['Dealer'].str.contains('TSO Mobile Peru')]
uts_sistemas_kalo=ubd[ubd['Dealer'].str.contains('UTS Sistemas Kalo MX')]
tso_mobile_colombia_ibo_idcom=ubd[ubd['Dealer'].str.contains('TSO Colombia - IBO IDCOM')]
tso_mobile_colombia_ibo_segtec=ubd[ubd['Dealer'].str.contains('TSO Colombia - IBO SEGTEC')]
tso_peru_fg_satelital_eirl=ubd[ubd['Dealer'].str.contains('TSO Peru -   FG SATELITAL EIRL')]
```

```python
col=pd.concat([tso_mobile_colombia,tso_mobile_colombia_ibo_idcom,tso_mobile_colombia_ibo_segtec])
per=pd.concat([tso_mobile_peru,tso_peru_fg_satelital_eirl])
mex=uts_sistemas_kalo
```

```python
total_fee_mex_per=per['Dec_Monthly_fee'].sum()+mex['Dec_Monthly_fee'].sum()
total_fee_col=col['Dec_Monthly_fee'].sum()
dollar=float(input('Digite el valor del dolar hoy: '))
mrr_latam=total_fee_col/dollar+total_fee_mex_per
print('Mrr LATAM: ',mrr_latam)
```
