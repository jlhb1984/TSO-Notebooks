```python
import pandas as pd
import numpy as np
```

```python
ubd_nf=pd.read_csv('/home/jlhb1984/py-projects/data/cases/CRMUnitsDealersReports_15724_20251202140152.csv',sep=';',encoding='latin1')
ubd=ubd_nf[ubd_nf['Last Event Date'].str.contains('2025')]
ubd_nf.info()
ubd.info()
```

```python
vec_aux=[]
for i in range(0,14578):
    aux=ubd.iloc[i,14].replace(',','.')
    vec_aux.append(float(aux))
ubd['Dec_Monthly_fee']=vec_aux

ubd.info()
#ubd.to_csv('/home/jlhb1984/py-projects/data/cases/mrr.csv')
```

```python
ubd['Dealer'].value_counts()
```

```python
tso_mobile_colombia=ubd[ubd['Dealer'].str.contains('TSO Mobile - Colombia')]
tso_mobile_peru=ubd[ubd['Dealer'].str.contains('TSO Mobile Peru')]
uts_sistemas_kalo=ubd[ubd['Dealer'].str.contains('UTS Sistemas Kalo MX')]
tso_mobile_colombia_ibo_idcom=ubd[ubd['Dealer'].str.contains('TSO Colombia - IBO IDCOM')]
tso_mobile_colombia_ibo_segtec=ubd[ubd['Dealer'].str.contains('TSO Colombia - IBO SEGTEC')]
tso_peru_fg_satelital_eirl=ubd[ubd['Dealer'].str.contains('TSO Peru -   FG SATELITAL EIRL')]
```

```python
col=pd.concat([tso_mobile_colombia,tso_mobile_colombia_ibo_idcom,tso_mobile_colombia_ibo_segtec])
per=pd.concat([tso_mobile_peru,tso_peru_fg_satelital_eirl])
mex=uts_sistemas_kalo
```

Calamp

```python
calamp_2630_col=col[col['Unit Type'].str.contains('2630')]
calamp_2630_col.info()
rows_calamp_2630_col=calamp_2630_col.shape[0]
calamp_2630_per=per[per['Unit Type'].str.contains('2630')]
calamp_2630_per.info()
rows_calamp_2630_per=calamp_2630_per.shape[0]
calamp_2630_mex=mex[mex['Unit Type'].str.contains('2630')]
calamp_2630_mex.info()
rows_calamp_2630_mex=calamp_2630_mex.shape[0]
calamp_2630_col.to_csv('/home/jlhb1984/py-projects/data/cases/col_mrr.csv')
```

```python
calamp_2630_col['Dec_Monthly_fee'].value_counts()
for i in range(0,rows_calamp_2630_col):
    if calamp_2630_col.iloc[i,28]>=5800000:
        calamp_2630_col.iloc[i,28]=calamp_2630_col.iloc[i,28]/100
#calamp_2630_col['Dec_Monthly_fee'].to_csv('/home/jlhb1984/py-projects/data/cases/col_mrr.csv')
(calamp_2630_col['Dec_Monthly_fee'].value_counts()).to_csv('/home/jlhb1984/py-projects/data/cases/col_mrr_calamp.csv')
```
