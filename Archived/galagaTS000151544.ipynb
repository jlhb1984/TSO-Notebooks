```python
import csv
prueba1={'Número de prueba':'','Número de geocercas':'','Archivo usado':'','Tiempo':'','Funcionó S/N':''}

num_prueba=input('Digitar el número de prueba: ')
prueba1.update({'Número de prueba':num_prueba})

num_geocerca=input('Digitar el número de geocercas: ')
prueba1.update({'Número de geocercas':num_geocerca})

archivo_usado=input("Digitar el archivo usado: ")
prueba1.update({'Archivo usado':archivo_usado})

tiempo=input('Digitar el tiempo en minutos: ')
prueba1.update({'Tiempo':tiempo})

resultado=input('Funcionó S/N? ')
prueba1.update({'Funcionó S/N':resultado})

prueba1

with open('/home/jlhb1984/py-projects/data/galaga/pruebas_geocercas.csv', mode='a', newline='') as file:
    file.write('')
    csv_writer = csv. DictWriter(file, fieldnames = prueba1.keys())
    csv_writer.writerow(prueba1)
```
